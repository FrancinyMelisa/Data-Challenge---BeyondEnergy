###**Efecto invernadero**

####El clima no ha sido constante a lo largo de los años, sin embargo, en los últimos años esta situación se ha agravado aún más. La ingeniera Vásquez nos brindó la siguiente definición sobre cambio climático: “cambio del clima esta atribuido directa e indirectamente a la actividad humana que altera la composición de la atmosfera global y que se suma a la variabilidad natural del clima observada durante periodos de tiempo comparables.” El efecto invernadero por su parte, y los gases que este produce, es lo que provoca el aumento de temperatura en el planeta. Seguidamente, veremos algunos de los principales sectores asociados a la generación de CO2 en Costa Rica. 

##### Tanto en la tabla como en la visualización a continuación, se ha completado un análisis que muestra la generación total de CO2 desde el año 1990 y hasta el 2017. La tabla incluye los siguientes sectores: energía, construcción y manufactura, transporte, combustibles y otros sectores. Los resultados que se pueden apreciar en las visualizaciones indican el gran aumento de emisiones a través de los años, y que en efecto a aumentado los gases del efecto invernadero.
